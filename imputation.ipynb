# Imputation of the dataset
Remove 0 in the dataset and replace with a valid value.

```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("dataset.csv")
```

Load dataset with many empty values

```python
df
```

###### Replace all 0 with NaN in order to work with fill functions

```python
df.replace(0, np.nan, inplace=True)
```

```python
df
```

###### Fill the dataset with the first valid observation forward the NaN

```python
df.fillna(method='bfill',inplace=True)
```

```python
 df.at[0, "idFrame"] = 0
```

```python
df
```

###### Save the dataset

```python
df.to_csv("datasetNoZero.csv")
```
